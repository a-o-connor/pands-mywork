```python
# Get a series of unique values from a column that contains some of the values in a delimited form
def get_unique_values(data_frame, name_of_col, delim = "/"):
    values_with_delims = data_frame[name_of_col].dropna().unique() # returns a numpy array of the unique values in the column
    unique_values = set([]) # initialise an empty set... #sets do not allow duplicates, only unique values
    for value in values_with_delims: #some of the items in the staff column are "Mark, Clark/Dean, McClean/Rod Mod", want to separate by the /
            value_as_list = re.split('/', value)
            unique_values.update(value_as_list)
    ds = pd.Series(list(unique_values)) # convert your set back to a list
    ds.sort_values() # sorts as alphabetical 
    return unique_values

get_unique_values(df, "Staff (delimited)")
```

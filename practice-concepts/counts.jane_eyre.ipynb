```python
# Create a string with all the upper case letters 
symbols = "ABCDEFGHIJKLMNOPQRSTUVWXYZ "
```

```python
# Read in the book
with open("jane_eyre.txt", "r") as b:
    book = b.read()

#Create it as one long string in upper case 
jane_shouting = book.upper()
```

```python
# 
counts = {
    s: jane_shouting.count(s) for s in symbols       
}

counts
```

```python
pairs_of_symbols = [a + b for a in symbols for b in symbols]
pairs_of_symbols
```

```python
count_pairs = {s: jane_shouting.count(s) for s in pairs_of_symbols}
count_pairs
```

```python
first_letter = "T"
```

```python
count_pairs["TA"]
```

```python
count_pairs[first_letter+"A"]
```

```python
for s in symbols:
    count_pairs[first_letter + s]
```

```python
freq_of_letter_after_t = {s: count_pairs[first_letter[-1]+s] for s in symbols}
freq_of_letter_after_t
```

```python
random.choices(list(count_pairs.keys()), weights = list(count_pairs.values()))
```

```python
sentence = "T"

for i in range(1,100): #run this 100 times to get a 100 letter sentence
    freq_of_letter_after_first_letter = {s: count_pairs[sentence[-1]+s] for s in symbols} #returns a dict object for each letter the frequency it occurs after T

    next = random.choices(list(freq_of_letter_after_first_letter.keys()), weights = list(freq_of_letter_after_first_letter.values()),k=1)[0]

    first_letter += next

first_letter
```
